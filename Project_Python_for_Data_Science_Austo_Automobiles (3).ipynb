```python
# Now we can check further for the missing values in Gender

# Gender: count and percentage (including missing values)

gender_counts = df['Gender'].value_counts(dropna=False)
gender_percent = df['Gender'].value_counts(normalize=True, dropna=False) * 100

# Passing both the variables in a Dictionary
gender_summary = {
    'Count': gender_counts,
    'Percentage': gender_percent
}

print(gender_summary)
```

```python
# Finding all the unique columns in the data.
df['Gender'].unique()
```







*   From the above operations we can say that, from this 4 categorical values, 2 of them ('Femal': 1, 'Femle': 1) are wrong entries. May be it's a type mistake when entering the data.

```python
# Replacing the values with 'Female'

df['Gender']= df['Gender'].replace({
    'Femal': 'Female',
    'Femle': 'Female'
})
```

```python
# Rechecking the missing values once again

gender_counts = df['Gender'].value_counts(dropna=False)
gender_counts
```
